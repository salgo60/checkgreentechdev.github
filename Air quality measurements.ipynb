```python
import urllib3, json
import pandas as pd 
http = urllib3.PoolManager() 
urlbase = "https://datavardluft.smhi.se/52North/api/v1/"
#/timeseries 
url = urlbase + "timeseries"
r = http.request('GET', url) 
data = json.loads(r.data.decode('utf-8')) 
```

```python
df = pd.DataFrame(data)  
df.head()
```
