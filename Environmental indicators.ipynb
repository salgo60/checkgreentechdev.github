### Environmental indicators
* [Environmental indicators](https://greentechdev.github.io/data/environmental-indicators/)

```python
import urllib3, json
import pandas as pd 
http = urllib3.PoolManager() 
urlbase = "https://indikatorer-api.naturvardsverket.se/api/v1/"
url = urlbase + "indicators"
r = http.request('GET', url) 
data = json.loads(r.data.decode('utf-8')) 
```

```python
df = pd.DataFrame(data)  
pd.set_option("display.max.rows", None)
df.head(10)
```
